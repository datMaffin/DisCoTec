```python
def getQoi(filename, col, mode):
    line = subprocess.check_output(['tail', '-1', filename])
    tmp = line.split()
    if mode == 2:
        return sqrt( float(tmp[col]) / float(tmp[0]) )
    else:
        return float(tmp[col]) / float(tmp[0])

def getQoiCombi(dir, col, mode):
    # read level vectors and coefficients from spaces dat
    levels = []
    coeffs = []
    for line in open(dir + 'ginstance/spaces.dat'):
        tmp = line.split()
        levels.append( [ int(x) for x in tmp[1:7] ] )
        coeffs.append(float(tmp[7]))
    
    #print levels
    #print coeffs
    
    # read nrg.dat files
    qois = []
    for i in range(0, len(coeffs) ):
        qois.append( getQoi( dir + '/ginstance' + str(i) + '/nrg.dat', col, mode ) )
            
    assert len(qois) == len(coeffs)
    
    #print qois
    
    # compute combi lambda
    combi_qoi = 0.0
    for i in range(0, len(qois)):
        combi_qoi += coeffs[i] * qois[i]
        
    #print combi_qoi
    
    return levels, qois, combi_qoi

def calcQoiError(refdir1, refdir2, combidir, col, mode ):
    if verbose:
        print combidir, refdir1, refdir2
    
    ref_lambda1 = getQoi( refdir1 + '/nrg.dat', col, mode )
        
    ref_lambda2 = getQoi( refdir2 + '/nrg.dat', col, mode )
        
    levels, lambdas, combi_lambda = getQoiCombi( combidir, col, mode )
    
    err1_lambdas = []
    err2_lambdas = []
    
    for i in range(0,len(levels)):
        err1_lambdas.append( abs(ref_lambda1 - lambdas[i]) )
        err2_lambdas.append( abs(ref_lambda2 - lambdas[i]) )
        if verbose:
            print str(levels[i][:-1]) + ' ' + str( lambdas[i]) + ' ' + str( abs(ref_lambda1 - lambdas[i]) ) + ' ' + str( abs(ref_lambda2 - lambdas[i]) )
        
    min_err1 = min( err1_lambdas )
    min_err2 = min( err2_lambdas )
        
    if verbose:
        print 'combi ' + str(combi_lambda) + ' ' + str( abs(ref_lambda1 - combi_lambda ) )  + ' ' + str( abs(ref_lambda2 - combi_lambda ) ) 
        print 'ref1 ' + str(ref_lambda1)
        print 'ref2 ' + str(ref_lambda2)
    
    err1_combi = abs(ref_lambda1 - combi_lambda )
    err2_combi = abs(ref_lambda2 - combi_lambda )
    
    return err1_combi, err2_combi, min_err1, min_err2
```

```python
def getTimeSeries(dir,col,nspec):
    filename = dir + '/nrg.dat'
            
    f = open(filename)

    t = []
    val = []

    i  = 0 
    for line in f:
        if (i % (nspec+1)) == 0:
            # even line number extract time
            t.append( float( line.split()[0] ) )
        if ((i-1) % (nspec+1)) == 0:
            # odd line number extract val of column col
            val.append( float( line.split()[col] ) )
            
        i += 1 
        
    return t,val
```
